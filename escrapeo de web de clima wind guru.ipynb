```python
from selenium import webdriver
import os
from time import sleep
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
import pandas as pd
import datetime
```

# A continuación, construimos las url necesarias para hacer el scraping. Se le agrega la fecha de todos los días del año al objeto "URL"

```python
url='https://www.wunderground.com/history/daily/ar/buenos-aires/SABE/date/'
lista_url=[]

for i in pd.Series(pd.date_range(start='2019-01-01', end='2019-07-31')).astype(str):
       lista_url.append(url+str(i))
```

# Comenzamos con el scrapeo

```python
#Creo un dataframe vacio y un contador

contador=0

df_empty=pd.DataFrame()

#creo lista vacia para incorporar las url de donde se extrae cada fila de información. 
#Luego va a ser útil para captar la fecha a la que corresponde cada fila.

lista_fechas=[]

## seteo las opciones del driver para que inicie el navegador expandiéndolo

option = webdriver.ChromeOptions()

option.add_argument('start-maximized')

option.add_argument('incognito')

## abro el navegador

browser = webdriver.Chrome(executable_path='C:/Users/Félix/Desktop/chromedriver', chrome_options=option)

sleep(2)

##indico la web desde dónde empezar

browser.get(lista_url[0])

sleep(3)

## cambio el formato de las unidades de medida (de F a Celcius)

browser.find_elements_by_xpath('//*[@id="wuSettings"]/i')[0].click()

sleep(1)

browser.find_elements_by_xpath('//*[@id="wuSettings-quick"]/div/a[2]')[0].click()

## for loop para buscar todas las URL de la lista y extraer todas las filas de la tabla del clima 
## Algunas tablas tienen registros con frecuencia menor a una hora. 

for url in lista_url:
    
    contador+=1
    
    print(contador)
    
    print(url)
    
    try:
        
        browser.get(url)
    
        sleep(2)
        
    except:
        
        browser.close()
        
    for i in range(1,31,1):
        
        try:
            
            df_empty=df_empty.append([browser.find_elements_by_xpath('//*[@id="history-observation-table"]/tbody/tr['+str(i)+']')[0].text.split(' ')])
            
            lista_fechas.append(browser.current_url)
            
        except: 
            
            print('hay menos de 30 registros en ese día')        
    
```

```python
info_clima=df_empty

info_clima=pd.concat([info_clima,(info_clima.iloc[:,1].str.split('\n', expand=True))], axis=1)

direccion_viento=info_clima.iloc[:,6].str.split('\n', expand=True).iloc[:,1].reset_index().iloc[:,1]

info_clima=info_clima.reset_index().drop('index', axis=1)

info_clima=pd.concat([info_clima,direccion_viento], axis=1)

wind_speed_mph=info_clima.iloc[:,6].str.split('\n', expand=True).iloc[:,2]



info_clima=pd.concat([info_clima,wind_speed_mph], axis=1)

condicion=info_clima.iloc[:,13].str.split(expand=True).iloc[:,1]

info_clima=pd.concat([info_clima,condicion], axis=1)

```

## Conversion temperaturas

```python
info_clima['temperatura']=info_clima['temperatura'].apply(lambda x:(int(x)-32)*(5/9))
```

```python
info_clima['dew_point']=info_clima['dew_point'].apply(lambda x:(int(x)-32)*(5/9))
```

## Conversion velocidad

```python
info_clima['wind_gust']=info_clima['wind_gust'].apply(lambda x: (float(x)*1.60934))
```

```python
info_clima['velocidad_viento']=info_clima['velocidad_viento'].fillna(0)
```

```python
info_clima['velocidad_viento']=info_clima['velocidad_viento'].apply(lambda x: (float(x)*1.60934))
```

```python
info_clima.columns=[[                    'fecha', 'fecha_hora_completa',                'hora',
                      'ampm',         'temperaturaC',  'temperatura_medida',
                 'dew_pointC', 'temperatura_medida2',             'humedad',
          'velocidad_medida',           'wind_gustKMH',   'velocidad_medida2',
                   'presion',       'medidapresion',     'direccionviento',
          'velocidad_vientoKMH',          'condicion1',          'condicion2',
        'condicion_completa']]
```

```python
info_clima=info_clima[[                    'fecha', 'fecha_hora_completa',                'hora',
                      'ampm',         'temperaturaC', 'dew_pointC',     'humedad',
           'wind_gustKMH',   'presion',       'medidapresion',     'direccionviento',
          'velocidad_vientoKMH',  'condicion_completa']]
```

## Indicamos cuando es fin de semana

```python
clima18['fecha_hora_completa']=pd.to_datetime(clima18['fecha_hora_completa'], format='%Y-%m-%d %H:%M:%S')
```

```python
clima18['dia_semana']=clima18['fecha_hora_completa'].apply(lambda x: x.strftime('%w'))
```

```python
clima18['dia_semana']=clima18['dia_semana'].astype(int)
```

```python
fin_semana=[]
for i in clima18['dia_semana']:
    if (i==0) or (i==1):
        fin_semana.append(1)
    else:
        fin_semana.append(0)
        
```

```python
clima18['fin_semana']=fin_semana
```
